## DATA TRANSFORMATION

```
from sklearn import preprocessing
```

```
le = preprocessing.LabelEncoder()
```

### Attributes that need to encoded

```
arr = ['Department', 'Ward_Type',"Hospital_type_code", 'Bed Grade', 'Type of Admission', 'Age', 'Stay']
```

### Categories and its labels

```
for i in arr:
    print("\n\t\t",i,"\t\t")
    unique_values = df[i].unique()
    le.fit(unique_values)
    temp =  le.transform(unique_values)
    for i in range(len(unique_values)):
        print(unique_values[i], " = ", temp[i])
```

### Label Encoding

```
le.fit(df["Stay"])
transformed = le.transform(df["Stay"])
df["Stay"] = transformed
```

```
le.fit(df["Age"])
age_transformed = le.transform(df["Age"])
df["Age"] = age_transformed
```

```
le.fit(df["Department"])
dept_transformed = le.transform(df["Department"])
df["Department"] = dept_transformed
```

```
le.fit(df["Ward_Type"])
wt_transformed = le.transform(df["Ward_Type"])
df["Ward_Type"] = wt_transformed
```

```
le.fit(df["Type of Admission"])
toa_transformed = le.transform(df["Type of Admission"])
df["Type of Admission"] = toa_transformed
```

```
le.fit(df["Severity of Illness"])
soi_transformed = le.transform(df["Severity of Illness"])
df["Severity of Illness"] = soi_transformed
```

```
le.fit(df["Hospital_type_code"])
htc_transformed = le.transform(df["Hospital_type_code"])
df["Hospital_type_code"] = htc_transformed
```

### Removing unwanted attributes

```
features = ['Hospital_code', "Hospital_type_code",'Department', 'Ward_Type','Bed Grade', 'Type of Admission','Visitors with Patient', 'Age', "Stay"]
df = df[features]
```

### PREDICTION MODEL

```
df[df["Hospital_code"] == 26]

# each hospital code have a same hospital type, citycode hospital, hospital_region
```

```
df.columns.values
```

```
features = ['Hospital_code', "Hospital_type_code",'Department', 'Ward_Type','Bed Grade', 'Type of Admission','Visitors with Patient', 'Age']
```

```
x = df[features]
y = df["Stay"]
```

```
test_data = pd.read_csv("/datasets/testdata/test_data.csv")
```

```
test_data.head()
```

```
test_data.shape
```

```
test_data.isna().sum()
```

```
test_data["Bed Grade"].fillna(2.0, inplace=True)
```

```
test_data.isna().sum()
```

```
test_data['Department']= le.fit_transform(test_data['Department'])
```

```
test_data['Ward_Type']= le.fit_transform(test_data['Ward_Type'])
```

```
test_data['Hospital_type_code']= le.fit_transform(test_data['Hospital_type_code'])
```

```
test_data['Age']= le.fit_transform(test_data['Age'])
```

```
test_data['Type of Admission']= le.fit_transform(test_data['Type of Admission'])
```

```
test_data.head()
```

```
from sklearn. ensemble import RandomForestClassifier
from sklearn .metrics import accuracy_score
rfc = RandomForestClassifier(n_estimators=150)
rfc. fit(x, y)
y_test_preds = rfc. predict(test_data[features])
accuracy = accuracy_score(y_test_preds, test_data["Stay"])
accuracy
```

```
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x, y)
y_pred = dtc.predict(test_data[features])
accuracy=accuracy_score(y_pred,test_data["Stay"])
accuracy
```

```
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(random_state=1, max_iter=1000).fit(x, y)
y_pred = clf.predict(test_data[features])
accuracy=accuracy_score(y_pred,test_data["Stay"])
accuracy
```

```
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnbmodel=gnb.fit(x,y)
gnbpred=gnb.predict(test_data[features])
accuracy=accuracy_score(gnbpred,test_data["Stay"])
accuracy
```
